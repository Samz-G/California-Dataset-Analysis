# **Exploratory Data Analysis and Data Preprocessing**

This is a project entailing Exploratory Data Analysis and Data Preprocessing. Exploratory Data Analysis is a very fundamental process in both data analysis and modelling. It is the stage where one seeks to understand the structure, quality and insights from the raw dataset without altering it.

Data modeling is the process of defining, organizing, and structuring data to represent relationships, patterns, and attributes within a dataset. It is a foundational stage in machine learning and data analysis where data is prepared, cleaned, and shaped to create a framework that supports further analysis or model development.

The objectives of EDA stage, is to assess the dataset for completeness, patterns, distributions, and potential issues.

It exhausts techniques such as summary statistics, data visualization(for identifying trends or anomalies), correlation analysis, and data quality checks.

# **Data Preprocessing and Analysis**

```python
columns_to_keep = ['longitude', 'latitude', 'housing_median_age'] # They don't have any outliers
dropped_columns = data[columns_to_keep]

#print the columns after  dropping
data = data.drop(columns = columns_to_keep)# We temporarily drop them that we may deal with the outliers in the other 6 columns

columns_to_transform = ['total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']
# The ones that we proved to be having some outliers

# Since we saw that the outliers exist in the upper bound region of the above 6 columns, we replace them with Q3 for each column
for col in columns_to_transform:
    Q3 = data[col].quantile(0.75)
    Q1 = data[col].quantile(0.25)

    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR

    # Replace outliers
    data[col] = np.where(data[col] > upper_bound, upper_bound, data[col])

# Since we still need all the columns even after dealing with the outliers, we merge back the temporarily dropped columns
data = pd.concat([dropped_columns, data], axis=1) # Even after the tranformations, we still refer to our dataframe as 'data'

# The following lines ensures that all columns are visible when printing
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print(data)
```
